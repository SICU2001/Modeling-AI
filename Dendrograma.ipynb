#**Tutorial**

#Librerías

```python
import plotly.figure_factory as ff
import numpy as np
```

# Dendograma básico

```python
np.random.seed(1)

X = np.random.rand(15, 12) # 15 samples, with 12 dimensions each
fig = ff.create_dendrogram(X)
fig.update_layout(width=800, height=500)
fig.show()
```

# Dendograma con umbral de color

```python
X = np.random.rand(15, 10) # 15 samples, with 10 dimensions each
fig = ff.create_dendrogram(X, color_threshold=1.5)
fig.update_layout(width=800, height=500)
fig.show()
```

# Dendograma con cambio de orientación y etiquetas

```python
X = np.random.rand(10, 12)
names = ['Jack', 'Oxana', 'John', 'Chelsea', 'Mark', 'Alice', 'Charlie', 'Rob', 'Lisa', 'Lily']
fig = ff.create_dendrogram(X, orientation='left', labels=names)
fig.update_layout(width=800, height=800)
fig.show()
```

#**Category of plants in iris dataset**

#Librerías

```python
import plotly.figure_factory as ff
import numpy as np
import pandas as pd
```

#Importación de datos

```python
irisdata = pd.read_csv('Iris.csv', index_col = 'Id')

print(irisdata.shape)
irisdata.head()
```

#Preprocesamiento de los datos

Dividir los datos en atributos y etiquetas

```python
X = irisdata.drop('Species', axis = 1)
#X=X.to_numpy()
names = irisdata['Species']
names=names.to_list()
```

# Dendograma

**Básico**

```python
fig = ff.create_dendrogram(X)
fig.update_layout(width=800, height=500)
fig.show()
```

**Con umbral de color**

```python
fig = ff.create_dendrogram(X, color_threshold=1.5)
fig.update_layout(width=2000, height=500)
fig.show()
```

**Con cambio de orientación y etiquetas**

```python
fig = ff.create_dendrogram(X, orientation='left', labels=names)
fig.update_layout(width=1000, height=2000)
fig.show()
```

# Diferencias de clasificación entre los dendrogramas y k-medias

Los dendrogramas tienen una manera de clasificar en la que se toma en cuenta la jerarquía, es un diagrama en el que se puede apreciar qué tan cercanas están dos subclasificaciones por la distancia requerida para unirlas. En cambio en k-medias las clasificación es plana y no se puede apreciar la relación entre cada uno de los diferentes grupos.

En k-media es necesario especificar el número de grupos o clusters y en dendrogramas no, aunque se pueden seleccionar diferentes métodos para realizar el dendrograma y en k-medias el método dado el número de grupos es invariable. Cuando la cantidad de datos es muy grande el dendrograma se vuelve muy difícil de interpretar y requiere mucho potencial computacional, en cambio k-medias es más escalable.
