#**Predict based on diagnostic measurements whether a patient has diabetes**

#Librerías

```python
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

#Importación y exploración de datos

```python
diabetesdata = pd.read_csv('diabetes.csv')

print(diabetesdata.shape)
diabetesdata.head()
```

```python
sns.pairplot(diabetesdata, hue='Outcome', palette='deep');
```

#Preprocesamiento de los datos

Dividir los datos en atributos y etiquetas

```python
X = diabetesdata.drop('Outcome', axis = 1)
Y = diabetesdata['Outcome']
```

Dividir los datos en training set y testing set

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.10)
```

#Entrenamiento





```python
svclassifier = SVC(kernel = 'rbf')
svclassifier.fit(X_train, Y_train)
```

#Evaluación del algoritmo

```python
Y_pred = svclassifier.predict(X_test)

print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```
